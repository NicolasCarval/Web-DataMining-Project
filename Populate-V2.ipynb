```python
dfmu=pd.read_json("musee.json")
print(len(dfmu))
dfmu.dtypes
```

```python
dfmu = dfmu[dfmu['Code Postal'].notna()]
dfmu["Code Postal"]=dfmu["Code Postal"].astype(int)
dfmu=dfmu.reset_index()
len(dfmu)
```

```python
from rdflib import Graph, Literal, RDF, URIRef
# rdflib knows about quite a few popular namespaces, like W3C ontologies, schema.org etc.
from rdflib.namespace import FOAF , XSD, RDFS

# Create a Graph
g = Graph()


baseURI="http://exemple.org"
place=URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#Place")
gare=URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#TrainStation")

for i in range(len(df)):
    start=URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#"+str(i))


    
# Add triples using store's add() method.
    g.add((start, RDF.type, gare))
    g.add((gare, RDFS.subClassOf, place))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#city"), Literal(df.loc[i,"city"],datatype=XSD.string)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#department"), Literal(df.loc[i,"NOM_DEP"],datatype=XSD.string)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#latitude"), Literal(df.loc[i,"Latitude"],datatype=XSD.float)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#longitude"), Literal(df.loc[i,"Longitude"],datatype=XSD.float)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#zipCode"), Literal(df.loc[i,"code_postal"],datatype=XSD.integer)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#name"), Literal(df.loc[i,"Nom_Gare"],datatype=XSD.string)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#numberOfUsers"), Literal(df.loc[i,"traffic"],datatype=XSD.string)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#region"), Literal(df.loc[i,"INSEE_REG"],datatype=XSD.integer)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#wifi"), Literal(df.loc[i,"service_wifi"],datatype=XSD.string)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#id_gare"), Literal(df.loc[i,"id_region"],datatype=XSD.string)))

museum=URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#Museum")
for i in range(len(df),len(dfmu)+len(df)):
    start=URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#"+str(i))
    j=i-len(df)
    g.add((start, RDF.type, museum))
    g.add((museum, RDFS.subClassOf, place))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#city"), Literal(dfmu.loc[j,"city"],datatype=XSD.string)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#department"), Literal(dfmu.loc[j,"Département"],datatype=XSD.string)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#latitude"), Literal(dfmu.loc[j,"Latitude"],datatype=XSD.float)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#longitude"), Literal(dfmu.loc[j,"Longitude"],datatype=XSD.float)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#zipCode"), Literal(dfmu.loc[j,"Code Postal"],datatype=XSD.integer)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#name"), Literal(dfmu.loc[j,"Nom officiel du musée"],datatype=XSD.string)))
    g.add((start, URIRef("http://www.semanticweb.org/sébastien/ontologies/2022/2/untitled-ontology-22#Localisation"), Literal(dfmu.loc[j,"Adresse"],datatype=XSD.string)))
    

# print all the data in the Notation3 format
print("--- printing mboxes ---")
print(g.serialize(format='ttl'))
g.serialize(destination='triple.ttl', format='turtle',base=baseURI)
```
