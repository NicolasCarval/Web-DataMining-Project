```python
import pandas as pd
import json 
```

```python
with open('Data/gares-equipees-du-wifi.json', 'r') as f:
    data = json.load(f)
    
with open('Data/frequentation-gares.json', 'r') as f:
    data2 = json.load(f)
```

```python
df = pd.DataFrame(data)

df2 = pd.DataFrame(data2)
```

```python
df.head()
```

```python
df2.head()
```

```python
df = pd.DataFrame(df.fields)
df2 = pd.DataFrame(df2.fields)
```

```python
df=pd.concat([df.drop(['fields'], axis=1), df['fields'].apply(pd.Series)], axis=1)
df.head()
```

```python
df2=pd.concat([df2.drop(['fields'], axis=1), df2['fields'].apply(pd.Series)], axis=1)
df2.head()
```

```python
df2["code_uic_complet"]=df2["code_uic_complet"].str[2:]
df2.head()
```

```python
df3 = df.merge(df2, left_on='uic', right_on='code_uic_complet')
```

```python
df3 = df3[["uic","nom_de_la_gare","nom_gare","code_postal","service_wifi","total_voyageurs_2019"]]
df3.head()
```

```python
df4 = pd.read_csv("Data/gares-tgv.csv",header=0,encoding="latin1")
df4.rename(columns={'Unnamed: 0': 'ID'},inplace=True)
df4.head()
```

```python
df3.describe()
```

```python
def func(x):
    if x < 10000:
        return "low"
    elif x < 50000:
        return "medium"
    else:
        return "high"

df3["traffic"] = df3["total_voyageurs_2019"].apply(func)
```

```python
df3.head()
```

```python
final = df4.merge(df3, left_on='Nom_Gare', right_on='nom_de_la_gare',how="left")
final2 = df4.merge(df3, left_on='Nom_Gare', right_on='nom_gare',how="left")
end = final
end
end.head()
```

```python
import math
final.head()
```

```python
for i in final["Nom_Gare"]:
    if(final.loc[final['Nom_Gare'] == i]["nom_gare"].isnull().values.any()):
            final.loc[final['Nom_Gare'] == i] = final2.loc[final2['Nom_Gare'] == i]
final.head()
final.dropna(inplace=True)
final.shape
```

```python
final = final[["Nom_Gare","INSEE_REG","NOM_REG","INSEE_DEP","NOM_DEP","Latitude","Longitude","code_postal","service_wifi","traffic"]]
final.head()
```
